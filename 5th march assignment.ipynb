```python
#1) How can you create a Bokeh plot using Python code?

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [3, 5, 1, 7, 2]

output_file("line.html")

fig = figure(title="Simple Line Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

fig.line(x, y, line_width=2)

show(fig)
```

```python
#2) What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

x = [1, 2, 3, 4, 5]
y = [3, 5, 1, 7, 2]

output_file("glyphs.html")

fig = figure(title="Glyph Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

fig.circle(x, y, size=20, fill_color="red", line_color="black")

fig.square(x, y, size=15, fill_color="blue", line_color="black")

fig.line(x, y, line_width=2, line_color="green")

show(fig)
```
